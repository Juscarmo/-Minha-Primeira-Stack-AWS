### Traduzindo os nomes das colunas

Primeiro, vamos definir um dicionário de mapeamento para as colunas.

```python
columns_translation = {
    'work_year': 'ano_de_trabalho',
    'experience_level': 'nivel_de_experiencia',
    'employment_type': 'tipo_de_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_do_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_do_empregado',
    'remote_ratio': 'percentual_remoto',
    'company_location': 'localizacao_da_empresa',
    'company_size': 'tamanho_da_empresa'
}

df = df.rename(columns=columns_translation)

print("Colunas do DataFrame traduzidas:")
display(df.columns)
```

### Traduzindo as categorias da coluna 'percentual_remoto'

```python
remote_translation = {
    0: 'Não remoto',
    50: 'Híbrido',
    100: 'Totalmente remoto'
}

df['percentual_remoto'] = df['percentual_remoto'].map(remote_translation)

print("Categorias da coluna 'percentual_remoto' traduzidas:")
display(df['percentual_remoto'].value_counts())
```

```python
import numpy as np
import pandas as pd

df_salarios = pd.DataFrame({
    "nome":["Ana","Paula","Davi","Lia","Luiz"],
    "salario":[4000, np.nan, 5000, np.nan, np.nan]
})
df_salarios["salario_medio+"] = df_salarios["salario"].ffill()
# Correção: Calcular a média arredondada e imprimi-la
mean_salario_rounded = df_salarios["salario"].mean().round(2)
print(f"Média do salário (arredondada): {mean_salario_rounded}")
# Correção: Exibir o DataFrame
df_salarios
```
